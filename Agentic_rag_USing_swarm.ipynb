```python
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from typing import List
import os
```

```python
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain_chroma import Chroma

embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
import os
from langchain.embeddings import OpenAIEmbeddings  # You can also use other embeddings like HuggingFaceEmbeddings
from langchain.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma

def pdf_vectorstore(folder_path: str):
    documents = []

    # Load documents from PDF and DOCX files
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if filename.endswith('.pdf'):
            loader = PyPDFLoader(file_path)
        elif filename.endswith('.docx'):
            loader = Docx2txtLoader(file_path)
        else:
            print(f"Unsupported file type: {filename}")
            continue
        documents.extend(loader.load())

    # Split documents into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(documents)

    # Specify the persist_directory using an absolute path
    persist_directory = os.path.abspath("./chroma_db")
    os.makedirs(persist_directory, exist_ok=True)

    # Initialize the vectorstore
    collection_name = "my_collection"
    vectorstore = Chroma.from_documents(
        collection_name=collection_name,
        documents=splits,
        embedding=embedding_function,
        persist_directory=persist_directory
    )
    print("vectorstore created")
    return vectorstore

# Load documents from a folder
folder_path = "/content/"  # Replace with the actual path to your folder
vectorstore=pdf_vectorstore(folder_path)
```

```python
from langchain.vectorstores.base import VectorStoreRetriever
def pdf_retriever(question):

  # Initialize the retriever
  retriever = VectorStoreRetriever(vectorstore=vectorstore)
  result=retriever.get_relevant_documents(question)

  return "\n\n".join(doc.page_content for doc in result)
```

```python
from swarm import Agent


def router_instruction(context_variables):
  """Router Function"""
  question = context_variables.get("question")
  if 'ESOP' in question:
    context_variables["route"] = 'vectorstore'
    return 'vectorstore'

  else:
    context_variables["route"] = 'websearch'
    return 'websearch'


# Define the Router Agent
Router_Agent = Agent(
    name='Router',
    instructions="""You are a specialized routing agent that determines whether to use a vectorstore or websearch to answer questions.

ROUTING RULES:
1. Use VECTORSTORE for:
   - Any questions containing 'ESOP' (Employee Stock Ownership Plan)
   - Questions about ESOP policies, procedures, or documentation
   - ESOP-related inquiries of any kind

2. Use WEBSEARCH for:
   - All other questions not related to ESOP
   - General knowledge queries
   - Current events
   - Any topic not specifically about ESOP

OUTPUT FORMAT:
You must respond with exactly one of these two options:
- 'vectorstore'
- 'websearch'

IMPORTANT:
- Respond only with the routing destination
- No explanation or additional text
- Be consistent in routing all ESOP-related queries to vectorstore
- When in doubt about ESOP relation, use websearch
- after response like vectorstore or websearch , call handoff to Retriever_Agent
""",
    # f"""
    # Route the user question to either a vectorstore or search the internet.
    # You are an expert at routing a user question to a vectorstore or web search.
    # Use the vectorstore for questions on concepts related to Retrieval-Augmented Generation.
    # You do not need to be stringent with the keywords in the question related to these topics.
    # Otherwise, use web-search.
    # intructions:{router_instruction}
    # """,
    # Route user questions to either vectorstore or web search.
    #         Use vectorstore for:
    #         - Questions about Retrieval-Augmented Generation (RAG)
    #         - Related concepts and implementations
    #         - ESOP-related queries

    #         Use web search for all other queries.,
    model=model,
)

# # Define the Router function
# def route_question(question):
#     # Simple routing logic based on keywords in the question
#     if "retrieval" in question.lower() or "augmentation" in question.lower():
#         context_variables["route"] = "vectorstore"
#     else:
#         context_variables["route"] = "websearch"
#     context_variables["question"] = question  # Store the question for later use
#     return context_variables["route"]

# # Attach the function to Router_Agent's functions list
# Router_Agent.functions.append(route_question)

# # Define the Retriever function
# def retrieve_answer(context_variables):
#     route=context_variables["route"]
#     if route == "vectorstore":
#         context_variables["retrieved_content"]=pdf_retriever(context_variables)
#         return context_variables["retrieved_content"]  # Assume pdf_retriever retrieves from vectorstore
#     else:
#       context_variables["retrieved_content"]=search_tool(context_variables)
#       return context_variables["retrieved_content"]    # Assume search_tool performs a web search


def retrieve_answer(context_variables):

    """
    Uses a prompt to determine whether to retrieve data from a vector store or perform a web search.
    """
    return f"""if selected route is "vectorstore" Retrieving content from vector store  use pdf_retriever else use search_tool(question)  """

# Define the Retriever Agent
Retriever_Agent = Agent(
    name="Retriever Agent",
    instructions=
    retrieve_answer,
    # f"""
    # Use the information retrieved from either the vectorstore or web search to answer the user's question.
    # You are an assistant for question-answering tasks. Use the retrieved context to provide a clear, concise answer.
    # yuo this instructions:{retrieve_answer}""",
    model=model,
)
Grader_Agent =  Agent(
  name='Answer Grader',
  instructions="""Filter out erroneous retrievals',
    "You are a grader assessing relevance of a retrieved document to a user question."
Based on the response from the retriever agent for the question evaluate whether the retrieved content is relevant to the question."

    Binary score 'yes' or 'no' score to indicate whether the document is relevant to the question"
    "You must answer 'yes' if the response from the 'Retriever_Agent' is in alignment with the question asked."
    "You must answer 'no' if the response from the 'Retriever_Agent' is not in alignment with the question asked."
    "Do not provide any preamble or explanations except for 'yes' or 'no'.""",
  model=model
)

# # Hallucination Grader: Checks factual grounding of the response
Hallucination_Grader = Agent(
    name="Hallucination Grader",
    instructions="""
        You are a hallucination grader ensuring that answers are grounded in factual content.
        Review the answer thoroughly to verify it aligns precisely with the question and is factually correct.
        Based on the response from the grader task for the question evaluate whether the answer is grounded in / supported by a set of facts."),
    Binary score 'yes' or 'no' score to indicate whether the answer is sync with the question asked"
    "Respond 'yes' if the answer is in useful and contains fact about the question asked."
    "Respond 'no' if the answer is not useful and does not contains fact about the question asked."
    "Do not provide any preamble or explanations except for 'yes' or 'no'.
    """,
    model=model
)
answer_grader = Agent(
    name="Answer Grader",
    instructions="""    Based on the response from the hallucination task for the question evaluate whether the answer is useful to resolve the question."
    "If the answer is 'yes' return a clear and concise answer."
    "If the answer is 'no' then perform a 'websearch' and return the response",
   "Return a clear and concise response if the response from 'hallucination_task' is 'yes'."
    "Perform a web search using 'search_tool' and return ta clear and concise response only if the response from 'Hallucination_Agent' is 'no'."
    "Otherwise respond as 'Sorry! unable to find a valid response'..
    """,
    model=model,  # Add transfer functions here
)
```

```python
def transfer_to_router_agent(**kwargs):
    """Transfer to Router Agent for routing user question to the appropriate retrieval method."""
    return Router_Agent

def transfer_to_retriever_agent(**kwargs):
    """Transfer to Retriever Agent for retrieving information from the selected source either vectorstore or websearch."""
    return Retriever_Agent

def transfer_to_grader_agent(**kwargs):
    """Transfer to Grader Agent to assess the relevance of the retrieved content."""
    return Grader_Agent

def transfer_to_hallucination_grader(**kwargs):
    """Transfer to Hallucination Grader for checking if the response is grounded and free from hallucinations."""
    return Hallucination_Grader

def transfer_to_answer_grader(**kwargs):
    """Transfer to Answer Grader to finalize the answer or rerun retrieval if needed."""
    return answer_grader

def transfer_to_triage(**kwargs):
    """Transfer to Triage Agent when managing the sequence of agents and coordinating each question's response path."""
    return Triage_Agent
```

```python
from swarm import Agent

Router_Agent.functions= [transfer_to_retriever_agent]
Retriever_Agent.functions=[pdf_retriever, search_tool,transfer_to_grader_agent]
Grader_Agent.functions=[transfer_to_hallucination_grader]
Hallucination_Grader.functions=[transfer_to_answer_grader]
answer_grader.functions=[transfer_to_triage]



def triage_intruction():

  return f"""

You are the orchestrator handling a sequence of agents to answer a user question. Store each agent's output in context_variables using the question as a unique key, passing outputs sequentially for grounded, relevant responses.
 Start with handoff the question to Router_Agent
 after getting route then  handoff to the Retriever_Agent.
 after getting information from Retriver_agent then handoff to the Grader_Agent.
 After getting Response from grader agent either "yes" or "no" the handoff to the Hallucination_Agent.
 After Getting response from Hallucination agnts handoff to the answer_grader
 after getting response handoff to the Triage_Agent
 for final answer generation

    """
# Triage Agent coordinates all agents in sequence
Triage_Agent = Agent(
    name="Triage Agent",
    instructions=triage_intruction(),
    # -> Grader -> Hallucination Grader -> Answer Grader.,
    model=model,
    functions=[
        transfer_to_router_agent,
        transfer_to_retriever_agent,
        transfer_to_grader_agent,
        transfer_to_hallucination_grader,
        transfer_to_answer_grader
    ]
)
```

```python


import json

from swarm import Swarm

def process_and_print_streaming_response(response):
    content = ""
    last_sender = ""

    for chunk in response:
        if "sender" in chunk:
            last_sender = chunk["sender"]

        if "content" in chunk and chunk["content"] is not None:
            if not content and last_sender:
                print(f"\033[94m{last_sender}:\033[0m", end=" ", flush=True)
                last_sender = ""
            print(chunk["content"], end="", flush=True)
            content += chunk["content"]

        if "tool_calls" in chunk and chunk["tool_calls"] is not None:
            for tool_call in chunk["tool_calls"]:
                f = tool_call["function"]
                name = f["name"]
                if not name:
                    continue
                print(f"\033[94m{last_sender}: \033[95m{name}\033[0m()")

        if "delim" in chunk and chunk["delim"] == "end" and content:
            print()  # End of response message
            content = ""

        if "response" in chunk:
            return chunk["response"]


def pretty_print_messages(messages) -> None:
    for message in messages:
        if message["role"] != "assistant":
            continue

        # print agent name in blue
        print(f"\033[94m{message['sender']}\033[0m:", end=" ")

        # print response, if any
        if message["content"]:
            print(message["content"])

        # print tool calls in purple, if any
        tool_calls = message.get("tool_calls") or []
        if len(tool_calls) > 1:
            print()
        for tool_call in tool_calls:
            f = tool_call["function"]
            name, args = f["name"], f["arguments"]
            arg_str = json.dumps(json.loads(args)).replace(":", "=")
            print(f"\033[95m{name}\033[0m({arg_str[1:-1]})")


def run_demo_loop(
    starting_agent, context_variables=None, stream=False, debug=False
) -> None:
    client = Swarm(ollama_client)
    # client = Swarm(llm_client)
    print("Starting Swarm CLI 🐝")

    messages = []
    agent = starting_agent

    while True:
        user_input = input("User Input: ")
        if user_input.lower() == "/exit":
          print("Exiting the loop. Goodbye!")
          break  # Exit the loop
        messages.append({"role": "user", "content": user_input})

        response = client.run(
            agent=agent,
            messages=messages,
            context_variables=context_variables or {},
            stream=stream,
            debug=debug,
        )

        if stream:
            response = process_and_print_streaming_response(response)
        else:
            pretty_print_messages(response.messages)

        messages.extend(response.messages)
        agent = response.agent

```

```python
question="How does exercise price determine for ESOP?"
context_variables={"question":question}

run_demo_loop(Triage_Agent, debug=True)
```
